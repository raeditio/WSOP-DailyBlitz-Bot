Download dataset from Roboflow

```python
from roboflow import Roboflow
from google.colab import userdata
rf = Roboflow(api_key=userdata.get('ROBOFLOW_API_KEY'))
project = rf.workspace("raeditio").project("poker-card-doocc-t6qgi")
version = project.version(5)
dataset = version.download("yolov11")
```

Train the YOLO11 model

```python
from ultralytics import YOLO
import os

# Load the nano YOLO11 model
model = YOLO('yolo11s.pt')

# Train the model using your downloaded Roboflow dataset
# Note: 'data.yaml' is the file inside the folder you download from Roboflow
yaml_path = os.path.join(dataset.location, "data.yaml")
# Start the training process
model.train(
    data=yaml_path,
    epochs=150,
    imgsz=1024,
    batch=16,
    device=0,             # Use GPU

    # Color augmentation
    hsv_h=0.01,           # Little hue change to ensure card colors are retained
    hsv_s=0.4,            # Saturation
    hsv_v=0.4,            # Brightness

    # No geometrical augmentation as card locations are fixed
    degrees=0.0, fliplr=0.0, flipud=0.0, perspective=0.0, mosaic=0.0, mixup=0.0,

    project='poker_bot',  # The parent folder where results will be saved
    name='yolo11_poker'   # The specific sub-folder for this training run
)

print("\n--- Training Complete! ---")
```
